```python

import geopandas as gpd 
# Load the shapefile with the complete set of necessary files
shapefile_path = './Samples1/Samples.shp'
samples = gpd.read_file(shapefile_path)

# Display the first few rows to confirm successful loading and examine the structure
samples.head()
```

```python
from sklearn.model_selection import train_test_split

# Shuffle and split the data
train, test_validation = train_test_split(samples, test_size=0.3, random_state=42)  # 70% training, 30% test+validation
test, validation = train_test_split(test_validation, test_size=0.5, random_state=42)  # Split the 30% into 15% test, 15% validation

# Saving the subsets into new shapefiles
train_path = './Samples1/training_set.shp'
test_path = './Samples1/testing_set.shp'
validation_path = './Samples1/validation_set.shp'

train.to_file(train_path)
test.to_file(test_path)
validation.to_file(validation_path)

(train_path, test_path, validation_path)
```

```python
import matplotlib.pyplot as plt

# Plotting the training, testing, and validation datasets
fig, ax = plt.subplots(figsize=(10, 10))

train.plot(ax=ax, edgecolor='blue', color='green', label='Training Set')
test.plot(ax=ax, edgecolor='red', color='orange', label='Testing Set')
validation.plot(ax=ax, edgecolor='green',color='yellow', label='Validation Set')

plt.legend()
plt.title('Split Shapefile into Training, Testing, and Validation Datasets')
plt.show()

```

```python
# Plotting the tile grids for training, testing, and validation datasets to visualize their distribution
fig, axs = plt.subplots(1, 3, figsize=(20, 6), sharex=True, sharey=True)

# Training set
train.plot(ax=axs[0], edgecolor='blue', facecolor='green')
axs[0].set_title('Training Set')

# Testing set
test.plot(ax=axs[1], edgecolor='red', facecolor='orange')
axs[1].set_title('Testing Set')

# Validation set
validation.plot(ax=axs[2], edgecolor='green', facecolor='yellow')
axs[2].set_title('Validation Set')

plt.suptitle('Splitted data into train , test and validation Datasets')
plt.show()
```
